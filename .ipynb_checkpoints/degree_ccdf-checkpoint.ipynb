```python
from pathlib import Path
import os
import csv
import random

import numpy as np
import matplotlib.pyplot as plt
import networkx as nx


# =========================
# 基本参数
# =========================

SEED = 42
random.seed(SEED)
np.random.seed(SEED)


# =========================
# 自动定位项目根目录
# =========================

BASE_DIR = Path.cwd()

while BASE_DIR.name != "network collapse" and BASE_DIR.parent != BASE_DIR:
    BASE_DIR = BASE_DIR.parent

if BASE_DIR.name != "network collapse" and (BASE_DIR / "network collapse").exists():
    BASE_DIR = BASE_DIR / "network collapse"

os.chdir(BASE_DIR)

DATA_DIR = BASE_DIR / "data"
REAL_DIR = DATA_DIR / "real_networks"
SBM_DIR = DATA_DIR / "sbm"
RESULT_DIR = BASE_DIR / "result"

OUT_DIR = RESULT_DIR / "degree_analysis_all"
CCDF_DIR = OUT_DIR / "ccdf"
RANK_DIR = OUT_DIR / "degree_rank"

DATA_DIR.mkdir(exist_ok=True)
REAL_DIR.mkdir(exist_ok=True)
SBM_DIR.mkdir(exist_ok=True)
RESULT_DIR.mkdir(exist_ok=True)
OUT_DIR.mkdir(exist_ok=True)
CCDF_DIR.mkdir(exist_ok=True)
RANK_DIR.mkdir(exist_ok=True)


# =========================
# 三种 SBM 网络
# 如果 data/sbm 里已有 gml，就直接读取；
# 如果没有，就自动生成并保存。
# =========================

SBM_DATASETS = {
    "sbm_strong": SBM_DIR / "sbm_strong.gml",
    "sbm_medium": SBM_DIR / "sbm_medium.gml",
    "sbm_weak": SBM_DIR / "sbm_weak.gml",
}


# =========================
# 所有真实网络
# 都指向文件夹，后面的读取函数会自动找 .gml 或 .mtx
# =========================

REAL_DATASETS = {
    "karate": REAL_DIR / "karate",
    "football": REAL_DIR / "football",
    "ca_netscience": REAL_DIR / "ca-netscience",
    "bio_diseasome": REAL_DIR / "bio-diseasome",
    "inf_USAir97": REAL_DIR / "inf-USAir97",
}


print("项目目录：", BASE_DIR)
print("数据目录：", DATA_DIR)
print("SBM 目录：", SBM_DIR)
print("真实网络目录：", REAL_DIR)
print("结果目录：", OUT_DIR)
print("NetworkX 版本：", nx.__version__)

print("\nSBM 网络：")
for name, path in SBM_DATASETS.items():
    print(name, "->", path, "存在吗：", path.exists())

print("\n真实网络：")
for name, path in REAL_DATASETS.items():
    print(name, "->", path, "存在吗：", path.exists())
```

```python
def find_first_graph_file(folder):
    """
    在一个网络文件夹里自动寻找可读取的图文件。
    支持：
    - .gml
    - .mtx
    - .edges
    - .txt
    """
    folder = Path(folder)

    candidates = []

    for suffix in ["*.gml", "*.mtx", "*.edges", "*.txt"]:
        candidates.extend(folder.rglob(suffix))

    candidates = [
        p for p in candidates
        if "readme" not in p.name.lower()
        and "label" not in p.name.lower()
        and "group" not in p.name.lower()
    ]

    candidates = sorted(candidates, key=lambda p: str(p).lower())

    if not candidates:
        raise FileNotFoundError(
            f"在 {folder} 下面没有找到 .gml/.mtx/.edges/.txt 文件"
        )

    return candidates[0]


def read_edge_list_like_file(path):
    """
    备用读取方式：
    如果某个 .mtx 文件不是标准 Matrix Market 格式，
    就把它当作边列表读取。
    """
    G = nx.Graph()

    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        lines = f.readlines()

    numeric_lines = []

    for line in lines:
        line = line.strip()

        if not line:
            continue

        if line.startswith("#") or line.startswith("%"):
            continue

        parts = line.replace(",", " ").split()

        nums = []
        for x in parts:
            try:
                nums.append(int(float(x)))
            except Exception:
                pass

        if len(nums) >= 2:
            numeric_lines.append(nums)

    if not numeric_lines:
        raise ValueError(f"{path} 里面没有读到可用边列表")

    # 有些非标准 mtx 第一行是：节点数 节点数 边数
    # 这种第一行不是边，要跳过
    first = numeric_lines[0]

    if len(first) >= 3:
        a, b, c = first[0], first[1], first[2]
        if a == b and c >= len(numeric_lines) - 5:
            numeric_lines = numeric_lines[1:]

    for nums in numeric_lines:
        u, v = nums[0], nums[1]
        if u != v:
            G.add_edge(u, v)

    return G


def load_simple_graph(path):
    """
    读取网络，并统一处理成：
    无向图、无权图、简单图、无自环、只取最大连通分量。
    """
    path = Path(path)

    # 如果传入的是文件夹，就自动寻找里面真正的图文件
    if path.is_dir():
        path = find_first_graph_file(path)

    if not path.exists():
        raise FileNotFoundError(f"找不到文件：{path}")

    suffix = path.suffix.lower()

    print("实际读取文件：", path)

    if suffix == ".gml":
        G_raw = nx.read_gml(path, label=None)

    elif suffix == ".mtx":
        from scipy.io import mmread
        from scipy import sparse

        try:
            A = mmread(str(path))

            if sparse.issparse(A):
                A = A.tocsr()
            else:
                A = sparse.csr_matrix(A)

            try:
                A.eliminate_zeros()
            except Exception:
                pass

            if hasattr(nx, "from_scipy_sparse_array"):
                G_raw = nx.from_scipy_sparse_array(
                    A,
                    create_using=nx.Graph()
                )
            else:
                G_raw = nx.from_scipy_sparse_matrix(
                    A,
                    create_using=nx.Graph()
                )

        except Exception as e:
            print("标准 Matrix Market 读取失败，改用边列表方式读取。")
            print("原始错误：", repr(e))
            G_raw = read_edge_list_like_file(path)

    elif suffix in [".edges", ".txt"]:
        G_raw = nx.read_edgelist(
            path,
            comments="#",
            nodetype=str,
            data=False
        )

    else:
        raise ValueError(f"暂不支持这个文件格式：{path}")

    # 统一成简单无向无权图
    G = nx.Graph()
    G.add_nodes_from(G_raw.nodes())
    G.add_edges_from(G_raw.edges())
    G.remove_edges_from(nx.selfloop_edges(G))

    # 只取最大连通分量
    if G.number_of_nodes() > 0 and not nx.is_connected(G):
        gcc_nodes = max(nx.connected_components(G), key=len)
        G = G.subgraph(gcc_nodes).copy()

    degrees = dict(G.degree())
    max_degree = max(degrees.values()) if degrees else 0
    avg_degree = sum(degrees.values()) / G.number_of_nodes() if G.number_of_nodes() > 0 else 0

    print("节点数：", G.number_of_nodes())
    print("边数：", G.number_of_edges())
    print("是否连通：", nx.is_connected(G) if G.number_of_nodes() > 0 else False)
    print("最大度：", max_degree)
    print("平均度：", avg_degree)
    print("-" * 60)

    return G


print("第 2 块加载完成：读取函数已准备好。")
```

```python
# =========================
# SBM 参数
# 8 个社区，节点数合计 200
# =========================

COMMUNITY_SIZES = [60, 40, 30, 25, 20, 12, 8, 5]


def make_sbm_probability_matrix(pin, pout, n_blocks):
    """
    构造 SBM 概率矩阵。
    对角线是 pin，非对角线是 pout。
    """
    P = []

    for i in range(n_blocks):
        row = []
        for j in range(n_blocks):
            if i == j:
                row.append(pin)
            else:
                row.append(pout)
        P.append(row)

    return P


SBM_CONFIGS = {
    "sbm_strong": {
        "pin": 0.28,
        "pout": 0.008,
    },
    "sbm_medium": {
        "pin": 0.20,
        "pout": 0.025,
    },
    "sbm_weak": {
        "pin": 0.14,
        "pout": 0.055,
    },
}


def generate_sbm_graph(name, config):
    """
    生成一个 SBM 网络，并给节点添加 block/community 属性。
    """
    n_blocks = len(COMMUNITY_SIZES)

    P = make_sbm_probability_matrix(
        pin=config["pin"],
        pout=config["pout"],
        n_blocks=n_blocks
    )

    G = nx.stochastic_block_model(
        sizes=COMMUNITY_SIZES,
        p=P,
        seed=SEED,
        directed=False,
        selfloops=False,
        sparse=True
    )

    # 添加社区属性，方便以后检查
    start = 0
    for cid, size in enumerate(COMMUNITY_SIZES):
        for node in range(start, start + size):
            G.nodes[node]["block"] = cid
        start += size

    G = nx.Graph(G)
    G.remove_edges_from(nx.selfloop_edges(G))

    return G


def load_or_generate_sbm(name, path):
    """
    如果 data/sbm 下已经有 SBM gml，就读取；
    如果没有，就生成并保存。
    """
    path = Path(path)

    if path.exists():
        print("=" * 70)
        print("读取已有 SBM：", name)
        G = load_simple_graph(path)
        return G

    print("=" * 70)
    print("没有找到已有 SBM，开始生成：", name)

    G = generate_sbm_graph(name, SBM_CONFIGS[name])

    nx.write_gml(G, path)

    print("SBM 已生成并保存：", path)
    print("节点数：", G.number_of_nodes())
    print("边数：", G.number_of_edges())
    print("-" * 60)

    return G


# =========================
# 读取所有网络
# =========================

graphs = {}

# 1. 三种 SBM
for name, path in SBM_DATASETS.items():
    graphs[name] = load_or_generate_sbm(name, path)

# 2. 所有真实网络
for name, path in REAL_DATASETS.items():
    print("=" * 70)
    print("读取真实网络：", name)
    graphs[name] = load_simple_graph(path)

print("=" * 70)
print("全部网络读取完成。当前 graphs 包含：")
for name, G in graphs.items():
    print(
        name,
        "节点数：", G.number_of_nodes(),
        "边数：", G.number_of_edges()
    )
```

```python
def degree_ccdf(G):
    """
    计算 degree CCDF：
    x = degree k
    y = P(K >= k)
    counts_ge_k = 度数大于等于 k 的节点数量
    """
    degrees = np.array([d for _, d in G.degree()], dtype=int)

    # 去掉 0 度点，避免 log 坐标不能显示
    degrees = degrees[degrees > 0]

    if len(degrees) == 0:
        return np.array([]), np.array([]), np.array([])

    unique_degrees = np.sort(np.unique(degrees))
    n = len(degrees)

    ccdf_values = []
    counts_ge_k = []

    for k in unique_degrees:
        count = np.sum(degrees >= k)
        counts_ge_k.append(count)
        ccdf_values.append(count / n)

    return unique_degrees, np.array(ccdf_values), np.array(counts_ge_k)


def plot_single_degree_ccdf(G, name, out_dir=CCDF_DIR):
    """
    给单个网络画 degree CCDF。
    使用 log-log 坐标。
    """
    x, y, counts = degree_ccdf(G)

    fig, ax = plt.subplots(figsize=(5.8, 4.2))

    ax.loglog(
        x,
        y,
        marker="o",
        linewidth=1.5,
        markersize=4
    )

    ax.set_xlabel("Degree k")
    ax.set_ylabel("CCDF P(K ≥ k)")
    ax.set_title(f"{name}: Degree CCDF, log-log scale")

    ax.grid(True, which="both", alpha=0.3)
    fig.tight_layout()

    fig_path = out_dir / f"{name}_degree_ccdf_loglog.png"
    fig.savefig(fig_path, dpi=250, bbox_inches="tight")
    plt.show()

    print("CCDF 图已保存：", fig_path)


# =========================
# 1. 保存所有网络的 CCDF 数据
# =========================

ccdf_csv_path = CCDF_DIR / "all_networks_degree_ccdf.csv"

with open(ccdf_csv_path, "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.DictWriter(
        f,
        fieldnames=[
            "network",
            "degree_k",
            "ccdf",
            "count_degree_ge_k",
            "num_nodes",
            "num_edges",
        ]
    )

    writer.writeheader()

    for name, G in graphs.items():
        x, y, counts = degree_ccdf(G)

        for k, ccdf_value, count_value in zip(x, y, counts):
            writer.writerow({
                "network": name,
                "degree_k": int(k),
                "ccdf": float(ccdf_value),
                "count_degree_ge_k": int(count_value),
                "num_nodes": G.number_of_nodes(),
                "num_edges": G.number_of_edges(),
            })

print("CCDF 数据已保存：", ccdf_csv_path)


# =========================
# 2. 每个网络单独画一张 CCDF 图
# =========================

for name, G in graphs.items():
    plot_single_degree_ccdf(G, name)


# =========================
# 3. 所有网络画在一张总对比图里
# =========================

fig, ax = plt.subplots(figsize=(8, 5.5))

for name, G in graphs.items():
    x, y, counts = degree_ccdf(G)

    if len(x) == 0:
        continue

    ax.loglog(
        x,
        y,
        marker="o",
        linewidth=1.5,
        markersize=3,
        label=name
    )

ax.set_xlabel("Degree k")
ax.set_ylabel("CCDF P(K ≥ k)")
ax.set_title("All networks: Degree CCDF comparison, log-log scale")

ax.grid(True, which="both", alpha=0.3)
ax.legend(fontsize=8)
fig.tight_layout()

fig_path = CCDF_DIR / "all_networks_degree_ccdf_loglog.png"
fig.savefig(fig_path, dpi=250, bbox_inches="tight")
plt.show()

print("CCDF 总对比图已保存：", fig_path)
```

```python
def degree_rank_sequence(G):
    """
    返回按度从大到小排列的 rank 和 degree。
    rank 从 1 开始。
    """
    degrees = sorted([d for _, d in G.degree()], reverse=True)
    ranks = np.arange(1, len(degrees) + 1)

    return ranks, np.array(degrees)


def plot_single_degree_rank(G, name, out_dir=RANK_DIR):
    """
    单个网络的 degree-rank 图：
    横轴：节点 rank
    纵轴：degree
    """
    ranks, degrees = degree_rank_sequence(G)

    fig, ax = plt.subplots(figsize=(5.8, 4.2))

    ax.plot(
        ranks,
        degrees,
        marker="o",
        linewidth=1.5,
        markersize=3
    )

    ax.set_xlabel("Node rank")
    ax.set_ylabel("Degree")
    ax.set_title(f"{name}: Degree-rank plot")

    ax.grid(True, alpha=0.3)
    fig.tight_layout()

    fig_path = out_dir / f"{name}_degree_rank.png"
    fig.savefig(fig_path, dpi=250, bbox_inches="tight")
    plt.show()

    print("Degree-rank 图已保存：", fig_path)


# =========================
# 1. 保存所有网络的 degree-rank 数据
# =========================

rank_csv_path = RANK_DIR / "all_networks_degree_rank.csv"

with open(rank_csv_path, "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.DictWriter(
        f,
        fieldnames=[
            "network",
            "rank",
            "degree",
            "num_nodes",
            "num_edges",
        ]
    )

    writer.writeheader()

    for name, G in graphs.items():
        ranks, degrees = degree_rank_sequence(G)

        for r, d in zip(ranks, degrees):
            writer.writerow({
                "network": name,
                "rank": int(r),
                "degree": int(d),
                "num_nodes": G.number_of_nodes(),
                "num_edges": G.number_of_edges(),
            })

print("Degree-rank 数据已保存：", rank_csv_path)


# =========================
# 2. 每个网络单独画 degree-rank 图
# =========================

for name, G in graphs.items():
    plot_single_degree_rank(G, name)


# =========================
# 3. 所有网络画在一张总对比图里
# =========================

fig, ax = plt.subplots(figsize=(8, 5.5))

for name, G in graphs.items():
    ranks, degrees = degree_rank_sequence(G)

    ax.plot(
        ranks,
        degrees,
        linewidth=1.5,
        label=name
    )

ax.set_xlabel("Node rank")
ax.set_ylabel("Degree")
ax.set_title("All networks: Degree-rank comparison")

ax.grid(True, alpha=0.3)
ax.legend(fontsize=8)
fig.tight_layout()

fig_path = RANK_DIR / "all_networks_degree_rank.png"
fig.savefig(fig_path, dpi=250, bbox_inches="tight")
plt.show()

print("Degree-rank 总对比图已保存：", fig_path)
```
